```python
## Determine if a year is a leap year using nested condition statement
year = int(input("Enter year : " ))

print(f"the entered year is {year}")
print(f"Checking whether {year} is a leap year.....")

if year%4 == 0 and year%100 != 0:
    print(f"{year} is a Leap year")
elif year%400 == 0:
    print(f"{year} is a Leap year")
else:
    print(f"{year} is not a Leap year")
```

```python
n = int(input("Enter Number: "))

# Store original number because temp will change
original = n
temp = n

count = 0
armstrong_sum = 0


# Count total digits
while temp > 0:
    count += 1
    temp //= 10


# Reset temp because it became 0
temp = n


# Calculate Armstrong sum
while temp > 0:

    # Extract last digit
    digit = temp % 10

    # Add digit^count to total
    armstrong_sum += digit ** count

    # Remove last digit
    temp //= 10


# Final comparison
if armstrong_sum == original:
    print("Armstrong Number")

else:
    print("Not Armstrong Number")
```
